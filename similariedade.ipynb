### Frequência do termo (TF - Term Frequency)

$TF(t, n) = \frac{f(t,n)}{|d|} $

com $f(t,n)$ sendo a frequência do termo $t$ no documento $d$ e $|d|$ a quantidade de termos no documento $d$.

```python
# import
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Transformando sentenças em vetores
vetor_tf = TfidfVectorizer(use_idf=False)
contagemPalavras = vetor_tf.fit_transform([sentenca_1, sentenca_2])
```

```python
# Criando Data Frame
DataFrame = pd.DataFrame(contagemPalavras.toarray(), columns=vetor_tf.get_feature_names_out())
DataFrame.index = ['Sentença 1', 'Sentença 2']
DataFrame
```

### Term Frequency-Inverse Document Frequency (TF-IDF)

$W_{i,j} = TF_{i,j} \times \log(\frac{N}{df_i})$



*   $TF_{i,j}$ = frequência do termo $i$ no documento $j$
*   $N$ = número total de documentos
*   $df_i$ = número de documentos que contém o termo $i$

[Explicação](https://mungingdata.wordpress.com/2017/11/25/episode-1-using-tf-idf-to-identify-the-signal-from-the-noise/)

```python
# Transformando sentenças em vetores
vetor_tfidf = TfidfVectorizer()
contagemPalavras = vetor_tfidf.fit_transform([sentenca_1, sentenca_2])
```

## Similaridade do coseno
[Material](https://www.sciencedirect.com/topics/computer-science/cosine-similarity#:~:text=Cosine%20similarity%20measures%20the%20similarity,document%20similarity%20in%20text%20analysis)

$cos(x, y) = \frac{x . y}{||x|| \times ||y||}$

$x.y \rightarrow$ Produto interno

$x = [1, 2, 3]$ e $y = [4, 5, 6]$

$x.y = 1 \times 4 + 2 \times 5 + 3 \times 6$

$x.y = 32$

$||x|| → $ norma, módulo ou magnitude de um vetor

$||x|| = \sqrt{1^2 + 2^2 + 3^2}$

$||x|| = 3.7416$

```python
contagemPalavras.toarray()
```

```python
# import numpy
import numpy as np
```

```python
# Produto interno
x = contagemPalavras.toarray()[0]
y = contagemPalavras.toarray()[1]
produto_interno = np.dot(x, y)
print(produto_interno)
```

```python
# norma
norma_x = np.linalg.norm(x)
norma_y = np.linalg.norm(y)
norma = norma_x * norma_y
print(norma)
```

```python
# Coseno de similaridade
coseno = produto_interno / norma
print(coseno)
```

```python
# scipy
from scipy.spatial.distance import cosine

print(1 - cosine(x, y))
```
